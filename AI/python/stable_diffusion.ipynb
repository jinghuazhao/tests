```python
from diffusers import StableDiffusionPipeline
import torch
# Load the pretrained stable diffusion model
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id)
pipe = pipe.to("cuda")  # Use GPU for faster inference
# Generate images from text: Generate an image from a text description:
# Define the text prompt
prompt = "A fantasy landscape with mountains, a river, and a castle in the background"
# Generate the image
image = pipe(prompt).images[0]
# Display the image
image.show()
```
